```python
df_train = pd.read_csv('/content/fraudTrain.csv')
df_test = pd.read_csv('/content/fraudTest.csv')
```

```python
label_encoders = {}
for column in df_train.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_train[column] = le.fit_transform(df_train[column])
    label_encoders[column] = le
```

```python
X = df_train.drop('is_fraud', axis=1)
y = df_train['is_fraud']
```

```python
df_train.head()
```

```python
X = pd.get_dummies(X,columns=['gender', 'category', 'state', 'job'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.7, test_size = 0.3,random_state = 42)
```

```python
scaler = StandardScaler()
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
X_train_cleaned = X_train.dropna()
y_train_cleaned = y_train[X_train_cleaned.index]
```

```python
rf_classifier.fit(X_train_cleaned, y_train_cleaned)
```

```python
y_pred_rf = rf_classifier.predict(X_test)
```

```python
print("Random Forest Classifier:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
```

```python
conf_mat = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Random Forest)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(np.sort(y_pred_rf))
plt.xlabel("Index")
plt.ylabel("Predicted Probability of Fraud")
plt.title("Line Plot of Predicted Fraud Probabilities (Random Forest)")
plt.show()
```
